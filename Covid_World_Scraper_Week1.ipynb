```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url="https://www.worldometers.info/coronavirus/" #Site to pull out worldwide Covid19 data
```

```python
"""
Request to a webpage and returning response code
Respose code 200 indicates successful connection
"""
response=requests.get(url)
print (response)
```

```python
soup=BeautifulSoup(response.content, "html.parser") #Creating an object of BeautifulSoup
```

```python
soup.title.text #Prints Title of the web page.
```

```python
#Displays summary or latest figures of worldwise Covid19 cases
divs=soup.find("div", class_ ="panel_flip")
for i in divs.find_all("div", class_="panel_front"):
    print(i.text)
```

```python
coronatable=soup.find_all("table") #Locates table(s) in web page 
```

```python
len(coronatable)
```

```python
co=coronatable[0] #Create an empty list
```

```python
#Fetching World Covid19 cases table content
country=[]
total_cases=[]
new_cases=[]
total_deaths=[]
new_deaths=[]
total_recovered=[]
active_cases=[]
critical_cases=[]
total_cases_per1M_pop=[]
deaths_per1M_pop=[]
total_tests=[]
tests_per1M_pop=[]
rows=co.find_all("tr")[9:-8]
for row in rows:
    col=row.find_all("td")
    country.append(col[0].text.strip())
    total_cases.append(col[1].text.strip())
    new_cases.append(col[2].text.strip())
    total_deaths.append(col[3].text.strip())
    new_deaths.append(col[4].text.strip())
    total_recovered.append(col[5].text.strip())
    active_cases.append(col[6].text.strip())
    critical_cases.append(col[7].text.strip())
    total_cases_per1M_pop.append(col[8].text.strip())
    deaths_per1M_pop.append(col[9].text.strip())
    total_tests.append(col[10].text.strip())
    tests_per1M_pop.append(col[11].text.strip())
print(country)
print(total_cases)
print(new_cases)
print(total_deaths)
print(new_deaths)
print(total_recovered)
print(active_cases)
print(critical_cases)
print(total_cases_per1M_pop)
print(deaths_per1M_pop)
print(total_tests)
print(tests_per1M_pop)
```

```python
#Create DataFrame from the fetched lists
df=pd.DataFrame(list(zip(country, total_cases,new_cases,total_deaths,new_deaths,total_recovered,active_cases,critical_cases,total_cases_per1M_pop,deaths_per1M_pop,total_tests,tests_per1M_pop)),
                columns=["Country","Total_Cases","New_Cases","Total_Deaths","New_Deaths","Total_Recovered","Active_Cases","Critical_Cases","Cases/1M_Pop","Deaths/1M_Pop","Total_Tests","Tests/1M_Pop"])
df
```

```python
#Exploring dataframe
df.dtypes
```

```python
#Replace missing data with Nan. What is r'^\s*$' and is it something we can used as is to locate any missing data?
df.replace(r'^\s*$', np.nan, regex=True, inplace=True)
df
```

```python
df.info()
df.isnull().sum()
```

```python
df.replace(np.nan,0, inplace=True) #Replace NaN with 0 to prepare data for further analysis.
df
```

```python
#Cleaning New Cases variable. How to create a funtion to traverse thru the dataframe instead of writing following code sevaral times?
new_cases1=[]
for i in df["New_Cases"]:
    s=str(i)
    s1=s.replace("+","")
    s2=s1.replace(",","")
    s3=int(s2)
    new_cases1.append(s3)
    
print(new_cases1)
```

```python
#Clean Total Cases column
total_cases1=[]
for i in df["Total_Cases"]:
    x=str(i)
    x1=x.replace(",","")
    x2=int(x1)
    total_cases1.append(x2)
    
print(total_cases1)
```

```python
#Updating DataFrame with cleaned data
df1["Total_Cases"]=total_cases1
df1["New_Cases"]=new_cases1
df1
```
